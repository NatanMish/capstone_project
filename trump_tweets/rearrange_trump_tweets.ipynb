```python
import pandas as pd
import numpy as np
import ndjson
```

```python
df=pd.read_csv('all_trump_tweets_11062020.csv')
```

```python
print(len(df))
```

```python
df.dropna(inplace=True)
```

```python
print(len(df))
```

```python
df.columns
```

```python
df.drop(['is_retweet'],1,inplace=True)
```

```python
len(df.drop_duplicates('text'))
```

```python
df.loc[:,'created_at']=pd.to_datetime(df['created_at'], errors='ignore')
```

```python
df.sort_values('created_at',inplace=True)
```

```python
list_of_retweets_positions=[i for i in df.index if df.loc[i,'text'][:4]=='RT @']
```

```python
list_of_retweets_positions
```

```python
df=df.loc[~df.index.isin(list_of_retweets_positions)]
```

```python
len(df)
```

```python
df.to_csv('trump_tweets_reducted.csv',index=False)
```
