```python
import pandas as pd
import seaborn as sns
from scipy.stats import norm
```

```python
vp_dow = pd.read_csv('extended_mcmc_trump_sp500_vol_before_2016.csv', index_col=0)
```

```python
sns.set(rc={'figure.figsize':(4.5,2)})
series = vp_dow.trade_topic*100
mean=series.mean()
median=series.median()
plt = sns.distplot(series, bins=10)
plt.axvline(mean, color='r', linestyle='--')
# plt.legend({'Mean':series,'Distribution':mean})
plt.set(xlabel='Percentages', ylabel = 'Probability', title='Topic: Trade Indicator: S&P 500 Transaction Volume')
```
